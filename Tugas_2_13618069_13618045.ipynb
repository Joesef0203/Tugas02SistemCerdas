```python
hb1['arrival_date'] = hb1[[
    'arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']].apply(
    lambda x: '{}-{}-{}'.format(x[0], x[1], x[2]), axis=1)
hb1['arrival_date'].head()
```

```python
loyal_cust = hb1[(hb1['arrival_date_year']==2016) & (hb1['is_repeated_guest'] ==1) & 
                 (hb1['tot_person']>4)]
loyal_cust.head()
```

```python
from sklearn.preprocessing import StandardScaler
variables= ['adults', 'children', 'babies']
x = hb1.loc[:, variables].values
y = hb1.loc[:, ['customer_type']].values
x = StandardScaler().fit_transform(x)
x = pd.DataFrame(x)
x.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
x_pca = pca.fit_transform(x)
x_pca = pd.DataFrame(x_pca)
x_pca.head()
```

```python
explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
x_pca['customer_type']=y
x_pca.columns = ['PC1','PC2','PC3','customer_type']
x_pca.head()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.set_xlabel('Principal Component1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3 Component PCA')
customer_types = ["Transient", "Group", "Transient-Party", "Contract"]
colors = ['r', 'b','g','y']
for customer_type, color in zip(customer_types,colors):
    IndicesToKeep = x_pca['customer_type'] == customer_type
    ax.scatter(x_pca.loc[IndicesToKeep, 'PC1']
              , x_pca.loc[IndicesToKeep, 'PC2']
               , x_pca.loc[IndicesToKeep, 'PC3']
              , c = color
              , s = 100)
    ax.legend(customer_types)
    ax.grid()
```
